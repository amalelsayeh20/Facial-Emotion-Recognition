```python
# INSTALL REQUIRED PACKAGES (RUN ONCE)
# !pip install opencv-python-headless pillow

# 1пёЏвѓЈ IMPORT LIBRARIES
import tensorflow as tf
import numpy as np
import base64
import io
from PIL import Image
import cv2
from google.colab import output
from IPython.display import HTML, display

# 2пёЏвѓЈ LOAD YOUR SAVED MODEL
MODEL_PATH = '/content/my_emotion_model.h5'
model = tf.keras.models.load_model(MODEL_PATH)

# Model expects grayscale 200x200 images (from your training config)
IMG_SIZE = 200
LABELS = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']

# 3пёЏвѓЈ DEFINE PREDICTION FUNCTION
def predict_from_webcam(image_data: str):
    """
    Process base64 image from webcam and return emotion prediction
    """
    try:
        # Decode base64 image
        decoded_data = base64.b64decode(image_data.split(',')[1])

        # Convert to grayscale PIL Image
        pil_image = Image.open(io.BytesIO(decoded_data)).convert('L')

        # Convert to numpy array and resize to model input size
        img_array = np.array(pil_image)
        resized = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))

        # Normalize (match your Rescaling layer)
        normalized = resized / 255.0

        # Add batch and channel dimensions [batch, height, width, channels]
        input_data = np.expand_dims(normalized, axis=(0, -1))

        # Make prediction
        predictions = model.predict(input_data, verbose=0)[0]

        # Get top emotion and confidence
        emotion_idx = np.argmax(predictions)
        emotion = LABELS[emotion_idx]
        confidence = predictions[emotion_idx]

        return f"{emotion} ({confidence:.1%})"

    except Exception as e:
        return f"Error: {str(e)}"

# 4пёЏвѓЈ REGISTER FUNCTION FOR COLAB JAVASBRIDGE
output.register_callback('predict_from_webcam', predict_from_webcam)

# 5пёЏвѓЈ CREATE INTERACTIVE CAMERA INTERFACE
html_interface = '''
<div style="font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif; max-width: 600px;">
  <h2 style="color: #2c3e50;">рџЋ­ Live Emotion Recognition</h2>

  <div style="margin-bottom: 15px;">
    <button id="startBtn" style="padding: 12px 24px; font-size: 16px; background: #27ae60; color: white; border: none; border-radius: 8px; cursor: pointer; margin-right: 10px;">
      в–¶пёЏ Start Camera
    </button>
    <button id="stopBtn" style="padding: 12px 24px; font-size: 16px; background: #e74c3c; color: white; border: none; border-radius: 8px; cursor: pointer; display: none;">
      вЏ№пёЏ Stop Camera
    </button>
  </div>

  <video id="video" autoplay style="border: 3px solid #3498db; border-radius: 10px; max-width: 100%; display: none;"></video>
  <canvas id="canvas" width="400" height="300" style="display: none;"></canvas>

  <div id="result" style="font-size: 28px; font-weight: bold; color: #2c3e50; margin-top: 20px; padding: 15px; background: #ecf0f1; border-radius: 8px; text-align: center;">
    Waiting for camera...
  </div>

  <div id="status" style="color: #7f8c8d; margin-top: 10px; font-size: 14px;"></div>
</div>

<script>
let stream = null;
let intervalId = null;

async function startCamera() {
  const video = document.getElementById('video');
  const canvas = document.getElementById('canvas');
  const result = document.getElementById('result');
  const status = document.getElementById('status');
  const startBtn = document.getElementById('startBtn');
  const stopBtn = document.getElementById('stopBtn');

  try {
    // Request camera access
    stream = await navigator.mediaDevices.getUserMedia({
      video: {
        width: { ideal: 400 },
        height: { ideal: 300 },
        facingMode: 'user'
      }
    });

    video.srcObject = stream;

    // Show elements
    video.style.display = 'block';
    startBtn.style.display = 'none';
    stopBtn.style.display = 'inline-block';
    status.textContent = 'вњ… Camera active - Analyzing emotions...';

    const ctx = canvas.getContext('2d');

    // Process frames every 200ms for smooth performance
    intervalId = setInterval(async () => {
      // Draw video frame to canvas
      ctx.drawImage(video, 0, 0, 400, 300);

      // Get JPEG image data (faster than PNG)
      const imageData = canvas.toDataURL('image/jpeg', 0.8);

      try {
        // Call Python prediction function
        const output = await google.colab.kernel.invokeFunction(
          'predict_from_webcam',
          [imageData],
          {}
        );

        // Update result display
        const prediction = output.data['text/plain'];
        result.textContent = 'рџЋ­ ' + prediction;

      } catch (err) {
        console.error('Prediction error:', err);
        status.textContent = 'вљ пёЏ Prediction error: ' + err.message;
      }
    }, 200); // 200ms = 5 FPS (adjustable)

  } catch (err) {
    console.error('Camera error:', err);
    status.textContent = 'вќЊ Camera error: ' + err.message + '. Please allow camera access and refresh the page.';
    stopCamera();
  }
}

function stopCamera() {
  const video = document.getElementById('video');
  const startBtn = document.getElementById('startBtn');
  const stopBtn = document.getElementById('stopBtn');
  const status = document.getElementById('status');
  const result = document.getElementById('result');

  // Stop frame processing
  if (intervalId) {
    clearInterval(intervalId);
    intervalId = null;
  }

  // Stop camera stream
  if (stream) {
    stream.getTracks().forEach(track => track.stop());
    stream = null;
  }

  // Hide video and reset UI
  video.style.display = 'none';
  startBtn.style.display = 'inline-block';
  stopBtn.style.display = 'none';
  status.textContent = 'Camera stopped';
  result.textContent = 'Waiting for camera...';
}

// Attach button event handlers
document.getElementById('startBtn').onclick = startCamera;
document.getElementById('stopBtn').onclick = stopCamera;

// Check if camera is available
navigator.mediaDevices.getUserMedia({video: true}).then(() => {
  document.getElementById('status').textContent = 'вњ“ Camera access granted. Click "Start Camera" to begin.';
}).catch(err => {
  document.getElementById('status').textContent = 'вќЊ Camera not available: ' + err.message;
});
</script>
'''

# 6пёЏвѓЈ DISPLAY THE INTERFACE
display(HTML(html_interface))
```
